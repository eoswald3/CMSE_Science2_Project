```python
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np
```

# Odeint

```python
def earth(w,t):
    func = np.zeros(6)
    func[0] = w[3]
    func[1] = w[4]
    func[2] = w[5]
    r = np.sqrt(w[0]**2 + w[1]**2 + w[2]**2)
    func[3] = -(G*m1*w[0])/r**3
    func[4] = -(G*m1*w[1])/r**3
    func[5] = -(G*m1*w[2])/r**3
    return func
```

```python
G = 4*np.pi**2
m1 = 1

x0 = 1
y0 = 0
z0 = 0
vx0 = 0
vy0 = 4
vz0 = 0

w = [x0,y0,z0,vx0,vy0,vz0]

t0 = 0
tf = 100
tau = .001

t = np.arange(t0,tf,tau)

sol = odeint(earth,w,t)

x = sol[:,0]
y = sol[:,1]
z = sol[:,2]
vx = sol[:,3]
vy = sol[:,4]
vz = sol[:,5]
```

```python
fig = plt.figure(figsize = (20,20))
ax = fig.add_subplot(111,projection='3d')

ax.plot(x,y,z)
```
